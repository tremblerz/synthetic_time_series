```python
import praw
import csv
import datetime
from datetime import datetime
import time
import requests
import pandas as pd
import re
import yfinance as yf
```

```python
stock = "GOOGL"
flag = False
stock = yf.Ticker(stock)
stock_history = stock.history(period="18y")
del stock_history["Dividends"]
del stock_history["Stock Splits"]
stock_history = stock_history.reset_index()

```

```python
stock_history.dtypes
```

```python
stock_history['Date'] = stock_history.apply(lambda row: datetime.fromtimestamp(row["Date"].value/(10**9)).strftime('%m/%d/%Y'), axis = 1)
```

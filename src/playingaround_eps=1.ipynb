```python
from synthcity.utils.datasets.time_series.pbc import PBCDataloader
from synthcity.utils.datasets.time_series.google_stocks import GoogleStocksDataloader
from synthcity.plugins.core.dataloader import TimeSeriesDataLoader
```

```python
# static_data, temporal_data, horizons, outcome = PBCDataloader().load()
static_data, temporal_data, horizons, outcome = GoogleStocksDataloader().load()
```

```python
data = TimeSeriesDataLoader(
    temporal_data=temporal_data,
    observation_times=horizons,
    static_data=static_data,
    outcome=outcome,
)
```

```python
Benchmarks.evaluate(
    # [('timegan', 'timegan', {"dp_enabled": False, "n_iter": 1000})],
    [
        ('DPtimeganEps1', 'timegan', {"mode": "DPRNN", "dp_enabled": True, "n_iter": 1000, "epsilon": 1.}),
        ('DPtimeganEps5', 'timegan', {"mode": "DPRNN", "dp_enabled": True, "n_iter": 1000, "epsilon": 20.})
    ],
    # [('DPtimeganEps10', 'timegan', {"mode": "DPRNN", "dp_enabled": True, "n_iter": 1000, "epsilon": 5.})]
    data,
    task_type='time_series',
    workspace='DPWorkspace'
)
```

```python
syn_model_with_dp = Plugins().get("timegan", mode="DPRNN", dp_enabled=True, n_iter=300, epsilon=1.)
```

```python
syn_model_with_dp.fit(data)
```

```python
syn_model_with_dp.plot(plt, data)
```
